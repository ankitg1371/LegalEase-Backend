```python
"""
Legal Documents Vector DB Ingestion Pipeline
Supports: Pinecone, Chroma, Qdrant, FAISS
"""

import json
import os
from typing import List, Dict, Any
from dataclasses import dataclass
import hashlib

# Choose your embedding provider (uncomment one)
# Option 1: OpenAI
# from openai import OpenAI

# Option 2: HuggingFace (free, local)
from sentence_transformers import SentenceTransformer

# Option 3: Cohere
# import cohere

@dataclass
class LegalChunk:
    """Structured chunk for legal documents"""
    id: str
    text: str
    metadata: Dict[str, Any]
    
class LegalDocumentChunker:
    """Smart chunking for legal documents preserving context"""
    
    def __init__(self, chunk_size: int = 1000, overlap: int = 200):
        self.chunk_size = chunk_size
        self.overlap = overlap
    
    def create_chunk_id(self, act_name: str, chapter: str, section: str, index: int) -> str:
        """Create unique, deterministic chunk ID"""
        content = f"{act_name}_{chapter}_{section}_{index}"
        return hashlib.md5(content.encode()).hexdigest()[:16]
    
    def chunk_section(self, act_name: str, act_number: str, chapter: Dict, 
                     section: Dict) -> List[LegalChunk]:
        """Chunk a section while preserving legal context"""
        chunks = []
        
        section_num = section.get('section_number', '')
        section_title = section.get('section_title', '')
        chapter_num = chapter.get('chapter_number', '')
        chapter_title = chapter.get('chapter_title', '')
        
        # Build full section text
        section_parts = []
        
        # Add section header
        header = f"Section {section_num}: {section_title}"
        section_parts.append(header)
        
        # Handle different section structures
        if 'text' in section:
            section_parts.append(section['text'])
        
        if 'subsections' in section:
            for subsection in section['subsections']:
                subsec_num = subsection.get('subsection_number', '')
                subsec_text = subsection.get('text', '')
                section_parts.append(f"({subsec_num}) {subsec_text}")
        
        if 'definitions' in section:
            for definition in section['definitions']:
                term = definition.get('term', '')
                defn = definition.get('definition', '')
                section_parts.append(f"{term}: {defn}")
        
        if 'key_definitions' in section:
            for definition in section['key_definitions']:
                term = definition.get('term', '')
                defn = definition.get('definition', '')
                section_parts.append(f"{term}: {defn}")
        
        if 'obligations' in section:
            section_parts.append("Key Obligations:")
            for obligation in section['obligations']:
                section_parts.append(f"- {obligation}")
        
        if 'rights' in section:
            section_parts.append("Rights:")
            for right in section['rights']:
                section_parts.append(f"- {right}")
        
        if 'duties' in section:
            section_parts.append("Duties:")
            for duty in section['duties']:
                section_parts.append(f"- {duty}")
        
        if 'penalty_schedule' in section:
            section_parts.append("Penalty Schedule:")
            for penalty in section['penalty_schedule']:
                breach = penalty.get('breach', '')
                penalty_amt = penalty.get('penalty', '')
                section_parts.append(f"- {breach}: {penalty_amt}")
        
        # Combine all parts
        full_text = "\n\n".join(section_parts)
        
        # Create chunks with overlap
        if len(full_text) <= self.chunk_size:
            # Single chunk
            chunk_id = self.create_chunk_id(act_name, chapter_num, section_num, 0)
            chunks.append(LegalChunk(
                id=chunk_id,
                text=full_text,
                metadata={
                    'act_name': act_name,
                    'act_number': act_number,
                    'chapter_number': chapter_num,
                    'chapter_title': chapter_title,
                    'section_number': section_num,
                    'section_title': section_title,
                    'chunk_index': 0,
                    'total_chunks': 1,
                    'doc_type': 'legal_section'
                }
            ))
        else:
            # Multiple chunks with overlap
            start = 0
            chunk_index = 0
            total_chunks = (len(full_text) - self.overlap) // (self.chunk_size - self.overlap) + 1
            
            while start < len(full_text):
                end = start + self.chunk_size
                chunk_text = full_text[start:end]
                
                # Include context header in each chunk
                if chunk_index > 0:
                    chunk_text = f"[Continued from {section_num}] " + chunk_text
                
                chunk_id = self.create_chunk_id(act_name, chapter_num, section_num, chunk_index)
                chunks.append(LegalChunk(
                    id=chunk_id,
                    text=chunk_text,
                    metadata={
                        'act_name': act_name,
                        'act_number': act_number,
                        'chapter_number': chapter_num,
                        'chapter_title': chapter_title,
                        'section_number': section_num,
                        'section_title': section_title,
                        'chunk_index': chunk_index,
                        'total_chunks': total_chunks,
                        'doc_type': 'legal_section'
                    }
                ))
                
                start = end - self.overlap
                chunk_index += 1
        
        return chunks
    
    def chunk_document(self, doc: Dict) -> List[LegalChunk]:
        """Chunk entire legal document"""
        all_chunks = []
        
        act_name = doc['act_name']
        act_number = doc['act_number']
        
        for chapter in doc['chapters']:
            for section in chapter.get('sections', []):
                chunks = self.chunk_section(act_name, act_number, chapter, section)
                all_chunks.extend(chunks)
        
        return all_chunks

class EmbeddingGenerator:
    """Generate embeddings using various providers"""
    
    def __init__(self, provider: str = 'openai', model: str = None):
        self.provider = provider
        
        if provider == 'openai':
            self.client = OpenAI(api_key=os.getenv('OPENAI_API_KEY'))
            self.model = model or 'text-embedding-3-small'
        
        elif provider == 'huggingface':
            # Free, runs locally
            from sentence_transformers import SentenceTransformer
            self.model = model or 'sentence-transformers/all-MiniLM-L6-v2'
            self.encoder = SentenceTransformer(self.model)
        
        elif provider == 'cohere':
            import cohere
            self.client = cohere.Client(os.getenv('COHERE_API_KEY'))
            self.model = model or 'embed-english-v3.0'
    
    def embed_batch(self, texts: List[str]) -> List[List[float]]:
        """Generate embeddings for batch of texts"""
        
        if self.provider == 'openai':
            # OpenAI API
            response = self.client.embeddings.create(
                input=texts,
                model=self.model
            )
            return [item.embedding for item in response.data]
        
        elif self.provider == 'huggingface':
            # Local HuggingFace
            embeddings = self.encoder.encode(texts)
            return embeddings.tolist()
        
        elif self.provider == 'cohere':
            # Cohere API
            response = self.client.embed(
                texts=texts,
                model=self.model,
                input_type='search_document'
            )
            return response.embeddings

class VectorDBIngester:
    """Ingest chunks into various vector databases"""
    
    def __init__(self, db_type: str, **config):
        self.db_type = db_type
        self.config = config
        self.client = None
        self._setup_db()
    
    def _setup_db(self):
        """Setup connection to vector DB"""
        
        if self.db_type == 'pinecone':
            from pinecone import Pinecone, ServerlessSpec
            
            pc = Pinecone(api_key=os.getenv('PINECONE_API_KEY'))
            
            index_name = self.config.get('index_name', 'legal-docs')
            dimension = self.config.get('dimension', 1536)
            
            # Create index if doesn't exist
            if index_name not in pc.list_indexes().names():
                pc.create_index(
                    name=index_name,
                    dimension=dimension,
                    metric='cosine',
                    spec=ServerlessSpec(
                        cloud='aws',
                        region='us-east-1'
                    )
                )
            
            self.client = pc.Index(index_name)
        
        elif self.db_type == 'chroma':
            import chromadb
            from chromadb.config import Settings
            
            persist_dir = self.config.get('persist_directory', './chroma_db')
            
            self.client = chromadb.Client(Settings(
                persist_directory=persist_dir,
                anonymized_telemetry=False
            ))
            
            collection_name = self.config.get('collection_name', 'legal_docs')
            self.collection = self.client.get_or_create_collection(
                name=collection_name,
                metadata={"hnsw:space": "cosine"}
            )
        
        elif self.db_type == 'qdrant':
            from qdrant_client import QdrantClient
            from qdrant_client.models import Distance, VectorParams
            
            url = self.config.get('url', 'http://localhost:6333')
            self.client = QdrantClient(url=url)
            
            collection_name = self.config.get('collection_name', 'legal_docs')
            dimension = self.config.get('dimension', 1536)
            
            # Create collection if doesn't exist
            try:
                self.client.get_collection(collection_name)
            except:
                self.client.create_collection(
                    collection_name=collection_name,
                    vectors_config=VectorParams(
                        size=dimension,
                        distance=Distance.COSINE
                    )
                )
            
            self.collection_name = collection_name
        
        elif self.db_type == 'faiss':
            import faiss
            import numpy as np
            
            dimension = self.config.get('dimension', 1536)
            
            # Create FAISS index with cosine similarity
            self.index = faiss.IndexFlatIP(dimension)
            self.index = faiss.IndexIDMap(self.index)
            
            # Store metadata separately for FAISS
            self.metadata_store = {}
    
    def ingest_batch(self, chunks: List[LegalChunk], embeddings: List[List[float]]):
        """Ingest batch of chunks with embeddings"""
        
        if self.db_type == 'pinecone':
            # Pinecone format
            vectors = []
            for chunk, embedding in zip(chunks, embeddings):
                vectors.append({
                    'id': chunk.id,
                    'values': embedding,
                    'metadata': {
                        **chunk.metadata,
                        'text': chunk.text[:1000]  # Store preview
                    }
                })
            
            self.client.upsert(vectors=vectors)
        
        elif self.db_type == 'chroma':
            # Chroma format
            ids = [chunk.id for chunk in chunks]
            texts = [chunk.text for chunk in chunks]
            metadatas = [chunk.metadata for chunk in chunks]
            
            self.collection.add(
                ids=ids,
                embeddings=embeddings,
                documents=texts,
                metadatas=metadatas
            )
        
        elif self.db_type == 'qdrant':
            from qdrant_client.models import PointStruct
            
            points = []
            for chunk, embedding in zip(chunks, embeddings):
                points.append(PointStruct(
                    id=chunk.id,
                    vector=embedding,
                    payload={
                        **chunk.metadata,
                        'text': chunk.text
                    }
                ))
            
            self.client.upsert(
                collection_name=self.collection_name,
                points=points
            )
        
        elif self.db_type == 'faiss':
            import numpy as np
            
            # Convert to numpy array
            embeddings_array = np.array(embeddings).astype('float32')
            
            # Normalize for cosine similarity
            faiss.normalize_L2(embeddings_array)
            
            # Create IDs (use hash of chunk.id)
            ids = np.array([int(chunk.id, 16) for chunk in chunks])
            
            # Add to index
            self.index.add_with_ids(embeddings_array, ids)
            
            # Store metadata
            for chunk in chunks:
                self.metadata_store[chunk.id] = {
                    **chunk.metadata,
                    'text': chunk.text
                }
    
    def save_index(self, path: str = './vector_index'):
        """Save index to disk (for FAISS)"""
        if self.db_type == 'faiss':
            import faiss
            faiss.write_index(self.index, f'{path}/faiss.index')
            
            with open(f'{path}/metadata.json', 'w') as f:
                json.dump(self.metadata_store, f, indent=2)

def main():
    """Main ingestion pipeline"""
    
    print("="*80)
    print("LEGAL DOCUMENTS VECTOR DB INGESTION PIPELINE")
    print("="*80)
    
    # 1. Load JSON
    print("\n[1/5] Loading legal documents JSON...")
    with open('legal_documents_rag.json', 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    # 2. Chunk documents
    print("[2/5] Chunking documents...")
    chunker = LegalDocumentChunker(chunk_size=1000, overlap=200)
    
    all_chunks = []
    for doc in data['legal_documents']:
        chunks = chunker.chunk_document(doc)
        all_chunks.extend(chunks)
        print(f"  - {doc['act_name']}: {len(chunks)} chunks")
    
    print(f"\nTotal chunks created: {len(all_chunks)}")
    
    # 3. Generate embeddings
    print("\n[3/5] Generating embeddings...")
    print("  Choose provider:")
    print("  1. OpenAI (best quality, paid)")
    print("  2. HuggingFace (free, local)")
    print("  3. Cohere (good quality, paid)")
    
    # For demo, using HuggingFace (free)
    embedder = EmbeddingGenerator(provider='huggingface')
    
    # Batch process embeddings
    batch_size = 32
    all_embeddings = []
    
    for i in range(0, len(all_chunks), batch_size):
        batch = all_chunks[i:i+batch_size]
        texts = [chunk.text for chunk in batch]
        embeddings = embedder.embed_batch(texts)
        all_embeddings.extend(embeddings)
        print(f"  Processed {min(i+batch_size, len(all_chunks))}/{len(all_chunks)} chunks")
    
    # 4. Setup vector DB
    print("\n[4/5] Setting up vector database...")
    print("  Choose vector DB:")
    print("  1. Pinecone (cloud, scalable)")
    print("  2. Chroma (local, easy)")
    print("  3. Qdrant (local/cloud, fast)")
    print("  4. FAISS (local, lightweight)")
    
    # For demo, using Chroma (easy local setup)
    db = VectorDBIngester(
        db_type='chroma',
        persist_directory='./legal_chroma_db',
        collection_name='legal_documents'
    )
    
    # 5. Ingest into DB
    print("\n[5/5] Ingesting into vector database...")
    batch_size = 100
    
    for i in range(0, len(all_chunks), batch_size):
        batch_chunks = all_chunks[i:i+batch_size]
        batch_embeddings = all_embeddings[i:i+batch_size]
        
        db.ingest_batch(batch_chunks, batch_embeddings)
        print(f"  Ingested {min(i+batch_size, len(all_chunks))}/{len(all_chunks)} chunks")
    
    print("\n" + "="*80)
    print("✓ INGESTION COMPLETE!")
    print("="*80)
    print(f"\nDatabase ready with {len(all_chunks)} legal document chunks")
    print("You can now query the database for legal information with semantic search")

if __name__ == '__main__':
    main()
```

```python
"""
LEGAL DOCUMENT INGESTION PIPELINE (Simplified)
Uses:
- ChromaDB (local vector database)
- SentenceTransformers (HuggingFace embeddings)
"""

import os
import json
import hashlib
from dataclasses import dataclass
from typing import List, Dict, Any
from sentence_transformers import SentenceTransformer
import chromadb
from chromadb.config import Settings


# ------------------------------------------------------------------------------
# 1️⃣ Data Classes
# ------------------------------------------------------------------------------
@dataclass
class LegalChunk:
    """Structured chunk for legal documents"""
    id: str
    text: str
    metadata: Dict[str, Any]


# ------------------------------------------------------------------------------
# 2️⃣ Chunking Logic
# ------------------------------------------------------------------------------
class LegalDocumentChunker:
    """Smart chunking for legal documents preserving context"""
    
    def __init__(self, chunk_size: int = 1000, overlap: int = 200):
        self.chunk_size = chunk_size
        self.overlap = overlap
    
    def create_chunk_id(self, act_name: str, chapter: str, section: str, index: int) -> str:
        """Create unique deterministic chunk ID"""
        content = f"{act_name}_{chapter}_{section}_{index}"
        return hashlib.md5(content.encode()).hexdigest()[:16]
    
    def chunk_section(self, act_name: str, act_number: str, chapter: Dict, section: Dict) -> List[LegalChunk]:
        """Chunk a single legal section"""
        chunks = []
        
        section_num = section.get('section_number', '')
        section_title = section.get('section_title', '')
        chapter_num = chapter.get('chapter_number', '')
        chapter_title = chapter.get('chapter_title', '')
        
        # Build full section text
        section_parts = [f"Section {section_num}: {section_title}"]
        
        for key in ['text', 'subsections', 'definitions', 'key_definitions', 'obligations', 'rights', 'duties', 'penalty_schedule']:
            if key not in section:
                continue
            
            if key == 'text':
                section_parts.append(section['text'])
            
            elif key in ['subsections', 'definitions', 'key_definitions']:
                for item in section[key]:
                    if isinstance(item, dict):
                        term = item.get('term', '') or item.get('subsection_number', '')
                        val = item.get('definition', '') or item.get('text', '')
                        section_parts.append(f"{term}: {val}")
            
            elif key in ['obligations', 'rights', 'duties']:
                section_parts.append(key.capitalize() + ":")
                section_parts.extend([f"- {x}" for x in section[key]])
            
            elif key == 'penalty_schedule':
                section_parts.append("Penalty Schedule:")
                for penalty in section['penalty_schedule']:
                    breach = penalty.get('breach', '')
                    penalty_amt = penalty.get('penalty', '')
                    section_parts.append(f"- {breach}: {penalty_amt}")
        
        full_text = "\n\n".join(section_parts)
        
        # Chunking with overlap
        if len(full_text) <= self.chunk_size:
            chunk_id = self.create_chunk_id(act_name, chapter_num, section_num, 0)
            chunks.append(LegalChunk(
                id=chunk_id,
                text=full_text,
                metadata={
                    'act_name': act_name,
                    'act_number': act_number,
                    'chapter_number': chapter_num,
                    'chapter_title': chapter_title,
                    'section_number': section_num,
                    'section_title': section_title,
                    'chunk_index': 0,
                    'total_chunks': 1,
                    'doc_type': 'legal_section'
                }
            ))
        else:
            start = 0
            chunk_index = 0
            total_chunks = (len(full_text) - self.overlap) // (self.chunk_size - self.overlap) + 1
            
            while start < len(full_text):
                end = start + self.chunk_size
                chunk_text = full_text[start:end]
                
                if chunk_index > 0:
                    chunk_text = f"[Continued from Section {section_num}] " + chunk_text
                
                chunk_id = self.create_chunk_id(act_name, chapter_num, section_num, chunk_index)
                chunks.append(LegalChunk(
                    id=chunk_id,
                    text=chunk_text,
                    metadata={
                        'act_name': act_name,
                        'act_number': act_number,
                        'chapter_number': chapter_num,
                        'chapter_title': chapter_title,
                        'section_number': section_num,
                        'section_title': section_title,
                        'chunk_index': chunk_index,
                        'total_chunks': total_chunks,
                        'doc_type': 'legal_section'
                    }
                ))
                
                start = end - self.overlap
                chunk_index += 1
        
        return chunks
    
    def chunk_document(self, doc: Dict) -> List[LegalChunk]:
        """Chunk an entire document"""
        all_chunks = []
        for chapter in doc.get('chapters', []):
            for section in chapter.get('sections', []):
                all_chunks.extend(self.chunk_section(doc['act_name'], doc['act_number'], chapter, section))
        return all_chunks


# ------------------------------------------------------------------------------
# 3️⃣ Embedding + ChromaDB Ingestion
# ------------------------------------------------------------------------------
class LegalDBIngestor:
    """Embed and store chunks into ChromaDB"""
    
    def __init__(self, persist_dir: str = "./legal_chroma_db", collection_name: str = "legal_documents"):
        self.persist_dir = persist_dir
        self.collection_name = collection_name
        
        # Initialize embedder
        self.embedder = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
        
        # Initialize Chroma client
        self.client = chromadb.PersistentClient(persist_dir)
        
        self.collection = self.client.create_collection(
            name=collection_name,
        )
    
    def embed_texts(self, texts: List[str]):
        return self.embedder.encode(texts, convert_to_numpy=True).tolist()
    
    def ingest(self, chunks: List[LegalChunk], batch_size: int = 100):
        """Batch ingest chunks"""
        for i in range(0, len(chunks), batch_size):
            batch = chunks[i:i+batch_size]
            texts = [c.text for c in batch]
            embeddings = self.embed_texts(texts)
            ids = [c.id for c in batch]
            metadatas = [c.metadata for c in batch]

            self.collection.add(
                ids=ids,
                embeddings=embeddings,
                documents=texts,
                metadatas=metadatas
            )
            print(f"  ✓ Ingested {min(i+batch_size, len(chunks))}/{len(chunks)} chunks")

        # 🔹 Force persistence
        # self.client.persist()
        print("🧠 Database persisted to disk.")
        
# ------------------------------------------------------------------------------
# 4️⃣ Main Pipeline (for Jupyter cell)
# ------------------------------------------------------------------------------
def run_ingestion(json_path: str = "legal_documents_rag.json"):
    """Run the ingestion pipeline"""
    print("⚖️ Starting Legal Document Ingestion Pipeline...\n")

    # 1. Load JSON
    with open(json_path, "r", encoding="utf-8") as f:
        data = json.load(f)
    documents = data.get("legal_documents", [])
    print(f"📘 Loaded {len(documents)} legal documents")

    # 2. Chunking
    chunker = LegalDocumentChunker(chunk_size=1000, overlap=200)
    all_chunks = []
    for doc in documents:
        chunks = chunker.chunk_document(doc)
        all_chunks.extend(chunks)
        print(f"  - {doc['act_name']}: {len(chunks)} chunks created")
    print(f"\nTotal chunks: {len(all_chunks)}")

    # 3. Embedding + Ingestion
    db = LegalDBIngestor()
    db.ingest(all_chunks, batch_size=64)
    
    print("\n✅ Ingestion Complete!")#OperationalError: no such column: collections.topic

    print(f"Database persisted at: {db.persist_dir}")
    print(f"Collection name: {db.collection_name}")


# ------------------------------------------------------------------------------
# Run the pipeline
# ------------------------------------------------------------------------------
run_ingestion("legal_documents_rag.json")
```

```python
# ─────────────────────────────────────────────────────────────
# ⚖️ Legal RAG System — ChromaDB + HuggingFace + Gemini
# ─────────────────────────────────────────────────────────────

import os
import json
from typing import List, Dict, Any, Tuple
from dataclasses import dataclass

# ───────────────────────────────
# 🔹 Dataclass for Search Results
# ───────────────────────────────
@dataclass
class SearchResult:
    text: str
    score: float
    metadata: Dict[str, Any]

    def citation(self) -> str:
        act = self.metadata.get("act_name", "")
        section = self.metadata.get("section_number", "")
        return f"Section {section} of {act}" if act or section else "Unknown Citation"

    def formatted(self) -> str:
        return f"""
Citation: {self.citation()}
Chapter: {self.metadata.get('chapter_title', '')}
Section: {self.metadata.get('section_title', '')}
Score: {self.score:.3f}

{self.text}

---
"""

# ───────────────────────────────
# ⚙️ Legal RAG Class
# ───────────────────────────────
class LegalRAGSystem:
    def __init__(self, persist_directory: str = "./legal_chroma_db", collection_name: str = "legal_documents"):
        # Setup embeddings
        from sentence_transformers import SentenceTransformer
        self.embedder = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

        # Setup Chroma
        import chromadb
        from chromadb.config import Settings

        self.client = chromadb.Client(Settings(persist_directory=persist_directory, anonymized_telemetry=False))
        self.collection = self.client.get_collection(collection_name)
        print(f"✅ Connected to Chroma collection: {collection_name}")

    # ───────────────────────────────
    # 🔸 Embedding a query
    # ───────────────────────────────
    def embed_query(self, query: str) -> List[float]:
        return self.embedder.encode([query])[0].tolist()

    # ───────────────────────────────
    # 🔸 Search in ChromaDB
    # ───────────────────────────────
    def search(self, query: str, top_k: int = 5, filters: Dict = None) -> List[SearchResult]:
        query_vec = self.embed_query(query)
        results = self.collection.query(query_embeddings=[query_vec], n_results=top_k, where=filters)

        return [
            SearchResult(
                text=results["documents"][0][i],
                score=1 - results["distances"][0][i],
                metadata=results["metadatas"][0][i],
            )
            for i in range(len(results["ids"][0]))
        ]

    # ───────────────────────────────
    # 🔸 Build retrieval context
    # ───────────────────────────────
    def build_context(self, query: str, top_k: int = 5) -> Tuple[str, List[str]]:
        results = self.search(query, top_k)
        context, citations = [], []

        for i, r in enumerate(results, 1):
            citations.append(r.citation())
            context.append(f"[Source {i}] {r.citation()}\n{r.text}\n")

        return "\n---\n".join(context), citations

    # ───────────────────────────────
    # 🔸 Generate answer (Gemini)
    # ───────────────────────────────
    def answer_query(self, query: str, use_llm: bool = True) -> str:
        context, citations = self.build_context(query)

        if not use_llm:
            return f"Query: {query}\n\n{context}\n\nSources:\n" + "\n".join(citations)

        import google.generativeai as genai
        genai.configure(api_key=os.getenv("GEMINI_API_KEY"))
        model = genai.GenerativeModel("gemini-2.5-flash")

        system_prompt = (
            "You are a legal assistant specializing in Indian IT and data protection law. "
            "Answer based only on the provided sections, with proper citations. "
            "If information is not in the context, say 'Not available in the provided context.'"
        )

        prompt = f"{system_prompt}\n\nQuestion: {query}\n\nLegal Provisions:\n{context}"
        response = model.generate_content(prompt)
        answer = response.text.strip() if hasattr(response, "text") else "No response from Gemini."

        return f"Query: {query}\n\nAnswer:\n{answer}\n\nSources:\n" + "\n".join(citations)
```
